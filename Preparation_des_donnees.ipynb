```python
#### 2.2 Temps de chargement :
import pandas as pd 
import time 
def mesurer_temps(file_csv, separateur = ";"):
    start_time = time.time() # Demare le chronometre
    df = pd.read_csv(file_csv, sep = separateur) # chargement csv
    end_time = time.time() # arret chronometre
    temps_total = end_time - start_time # calcul du temps ecoulé
    print(f"Le chargement du fichier a pris {temps_total:.4f} secondes.")
    return df

df = mesurer_temps('bod_livre124/NotebookJupyter/f04/sejour_hote.csv')
df.head()
```

```python
dotation = pd.read_csv('bod_livre124/bod_livre124/NotebookJupyter/f04/dotation_du_personnel.csv', sep= ";", encoding= "utf-8")
dotation.info()
```

##### Transforme sur la colonne dotation qui doit etre en float et non en object

```python
## remplacer les virgules par des points afin de bien preparer a la conversion numeric
dotation['dotation'] = dotation['dotation'].str.replace(',', '.')

## convertir en float 
dotation['dotation'] = pd.to_numeric(dotation['dotation'], errors = 'coerce')
dotation['dotation'].dtype
```

```python
### verification de l'exactitude des colonnes
dotation.dtypes
```

##### Lignes ou les noms sont manquants.

```python
dotation[dotation['nom'].isna()]
```

```python
dotation[dotation['nom'].isna() & dotation['prenom'].isna()]
```

##### Supprimer ces lignes

```python
dotation = dotation.dropna(subset=['nom', 'prenom'], how = 'all')
print(dotation)
```

```python
#### reindexer tout en gardant l'ancien indexation
### dotation.reset_index(drop=False)
#### reindexer en supprimant lancien indexation
dotation= dotation.reset_index(drop=True)
```

##### Traitement des NAN

```python
## remplacer les noms ou prenoms NAN par 'non_defini'
dotation['nom'] = dotation['nom'].fillna('non_defini')
dotation['prenom'] = dotation['prenom'].fillna('non_defini')
#print(dotation)
```

```python
## remplacer les dotations nan par la moyenne des dotations disponibles
dotation['dotation'] = dotation['dotation'].fillna(dotation['dotation'].mean())
dotation['dotation'] = dotation['dotation'].round(2)
print(dotation)
```

##### Sauvegarde 

```python
dotation.to_csv('dotation_du_personnel_nettoyer_new.csv', index= False, sep=";")
print("sauvegarde effectué")
```
