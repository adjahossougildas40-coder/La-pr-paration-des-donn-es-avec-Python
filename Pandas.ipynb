```python
data = pd.DataFrame({
    "service": ["IT", "IT", "RH", "RH", "RH"],
    "absence": ["Maladie", "Congé", "Maladie", "Congé", "Maladie"],
    "jours": [2, 1, 3, 2, 1]
})
data
```

##### groupby simple (1 colonne)
➕ Somme par service

```python
# total des jours par service
data.groupby('service')['jours'].sum()
```

```python
## duree moyenne des
data.groupby('service')["jours"].mean()
```

```python
### plusieurs aggregations
data.groupby("service").agg(
    total_jours = ("jours", "sum"),
    moyenne_jours = ("jours", "mean"),
    nbre_absences = ("jours", "count")
)
```

```python
## groupby + filtre
data.groupby("service").filter(lambda x: x["jours"].sum() > 3)
```
